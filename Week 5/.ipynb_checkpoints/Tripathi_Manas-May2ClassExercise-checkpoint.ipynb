```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy.linalg as nlp
```

```python
#l1 norm
arr = [1,2,3]
val = 0
for item in arr:
    val += abs(item)

print(val)
```

```python
#l2 norm
arr = [1,2,3]
val = 0
for item in arr:
    val += abs(item)**2

print((val)**(1/2))
```

```python
#l3 norm
arr = [1,2,3]
val = 0
for item in arr:
    val += (item)**3

print((val)**(1/3))
```

```python
#l4 norm
arr = [1,2,3]
val = 0
for item in arr:
    val += abs(item)**4

print((val)**(1/4))
```

```python
#l5 norm
arr = [1,2,3]
val = 0
for item in arr:
    val += abs(item)**5

print((val)**(1/5))
```
